```python
import numpy as np
import pandas as pd
from utils import load_data, preprocess_data
from config import INPUT_TEST_CSV
```

```python
df = load_data(INPUT_TEST_CSV)
x = df[df['video_id'] == 0]
y = df[df['video_id'] == 1]
y_renamed = y.rename(
    columns={col: f"{col}_mirror" for col in y.columns if col not in ["frame", "video_id"]}
)

# Merge on frame
merged = pd.merge(x, y_renamed, on="frame")
cols_to_keep = ["frame"]

# For x (original)
joints = ["shoulder","ankle"]
sides = ["left", "right"]
coords = ["x", "y"]

for side in sides:
    for joint in joints:
        for coord in coords:
            cols_to_keep.append(f"{side}_{joint}_{coord}")
            cols_to_keep.append(f"{side}_{joint}_{coord}_mirror")

# Filter merged dataframe
final = merged[cols_to_keep]

final
```
